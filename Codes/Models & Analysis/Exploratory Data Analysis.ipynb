```python
# Exploratory Data Analysis
import pandas as pd
import numpy as np
from sklearn import linear_model
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df_final_1 = pd.read_csv('/Users/LindaSong/Desktop/cleaned data/merged data 14 sim.csv')

# saving only the changed year data:
df_final_1 = df_final_1[df_final_1['LSI changed'] == 1]

# selecting the variables:
all_variables = ['Age dependency ratio', 'IMPORT FROM CHINA','Employment Share Agriculture', 'Employment Share Service','FDI Net Inflow', 'GDP', 'Capital Formation','Labor Force Participation', 'Net Migration','relative price of investment','Labor share of income' ]
columns_to_take_log = ['Age dependency ratio','IMPORT FROM CHINA','Employment Share Service', 'GDP', 'Labor Force Participation','Capital Formation','relative price of investment', 'Labor share of income']

# Filtering out rows with empty strings ('') or '..' values
filtered_df = df_final_1[(df_final_1 != '') & (df_final_1 != '..')].dropna()
filtered_df = filtered_df[all_variables]
filtered_df[all_variables].astype(float)
filtered_df = filtered_df.astype(float)
print(filtered_df.head())

# preprocess labor share of income again
filtered_df['Labor share of income'] = 100 * filtered_df['Labor share of income']

# selecting the variables to take logs:
columns_to_take_log = ['Age dependency ratio','IMPORT FROM CHINA','Employment Share Service', 'GDP', 'Labor Force Participation','Capital Formation','relative price of investment']
filtered_df[columns_to_take_log] = np.log(filtered_df[columns_to_take_log])

# Create the default pairplot
sns_plot = sns.pairplot(filtered_df)
# Show the plot
plt.show()









```
